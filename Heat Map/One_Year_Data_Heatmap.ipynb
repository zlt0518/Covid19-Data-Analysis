# Heat Map on the tweets data

This notebook would visualize the heatmap of all the covid tweets in a daily basis through folium.

```python
import pandas as pd
total_df = pd.read_csv('/home/zhoulita/john-litao-zhou-file/covid_data_unprocessed/dataFrame/tweet_df_loc_1.csv',low_memory=False,lineterminator='\n')
total_df['day_index'] = 1

for i in range(2,351):
    file_dir = '/home/zhoulita/john-litao-zhou-file/covid_data_unprocessed/dataFrame/tweet_df_loc_'+str(i)+'.csv'
    temp_df = pd.read_csv(file_dir,lineterminator='\n',low_memory=False)
    temp_df['day_index'] = i
    total_df = pd.concat([total_df,temp_df])
```

```python
total_df
```

```python
#define the boundary for the grid
lower_left_coordinates =[-125.000000,24.000000]
upper_right_coordinates =[-66.000000,50.000000]

filter_df = total_df[(total_df['Lontitude']>=lower_left_coordinates[0])&(total_df['Lontitude']<=upper_right_coordinates[0])&(total_df['Latitude']>=lower_left_coordinates[1])&(total_df['Latitude']<=upper_right_coordinates[1])]
filter_df.to_csv('/home/zhoulita/scratch/covid_data_unprocessed/dataFrame/tweet_df_loc_us.csv')
```

```python
import folium
from folium.plugins import HeatMapWithTime

#initialize the map and put the data into the map
map_year_heatmap = folium.Map(location=[36.648693, -95.8974445],max_bounds=True, zoom_start=4)

# List comprehension to make out list of lists
heat_data = [[[row['Latitude'],row['Lontitude']] for index, row in total_df[total_df['day_index'] == i].iterrows()] for i in range(1,351)]
# Plot it on the map
hm = HeatMapWithTime(heat_data,auto_play=True,max_opacity=0.8)
hm.add_to(map_year_heatmap)
map_year_heatmap
                                                                       
```
