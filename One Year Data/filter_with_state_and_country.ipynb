This notebook filter out the number of tweets with different states in US and the states of four English-spoken countries

```python
import pandas as pd
```

```python
total_df = pd.read_csv('/home/zhoulita/scratch/covid_data_unprocessed/dataFrame/tweet_df_loc_total.csv',low_memory=False,lineterminator='\n')
del total_df['Unnamed: 0']
del total_df['Unnamed: 0.1']

def get_week_index(df):
    return int((df['day_index']-1)/7)
total_df['week_index']= total_df.apply(get_week_index,axis = 1)
```

```python
us_pls_df = pd.read_csv("us_places.csv")
del us_pls_df['Unnamed: 0']
del us_pls_df['full_name']
del us_pls_df['place_type']
del us_pls_df['country_code']
del us_pls_df['country']
del us_pls_df['geo']
del us_pls_df['name']
del us_pls_df['Lontitude']
del us_pls_df['Latitude']
```

```python
df_merge = pd.merge(total_df,us_pls_df,how = 'inner',left_on = 'geo_id',right_on = 'id')
```

```python
df_merge
```

```python
week_data = df_merge[df_merge['week_index'] == 0]
state_data_week = pd.DataFrame(week_data['States'].value_counts())
state_data_week.reset_index(inplace=True)
state_data_week.rename(columns={'index':'States','States':'Count'},inplace=True)
state_data_week['Count'] = state_data_week['Count']/7
state_data_week['Week_index'] = 0
```

```python
for i in range(1,50):
    week_temp = df_merge[df_merge['week_index'] == i]
    temp = pd.DataFrame(week_temp['States'].value_counts())
    temp.reset_index(inplace=True)
    temp.rename(columns={'index':'States','States':'Count'},inplace=True)
    temp['Count'] = temp['Count']/7
    temp['Week_index'] = i
    state_data_week = pd.concat([state_data_week,temp])
```

```python
state_data_week.to_csv('us_states_weekly_counts.csv')
```
