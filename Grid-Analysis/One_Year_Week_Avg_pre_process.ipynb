# Preprocess the data toward the average of week for one year using the grid

This notebook is done on ComputeCanada and was used to load all the data of the tweets in the 70*35 grid in US, south of Canada and north of Mexico. The tweets with location id is filtered in to the grid. A table contains the average number of tweets of each gridin different weeks(0-50 week) is generated. 

```python
import pandas as pd
total_df = pd.read_csv('tweet_df_loc_1.csv',low_memory=False,lineterminator='\n')
total_df['day_index'] = 1

for i in range(2,351):
    file_dir = 'tweet_df_loc_'+str(i)+'.csv'
    temp_df = pd.read_csv(file_dir,lineterminator='\n',low_memory=False)
    temp_df['day_index'] = i
    total_df = pd.concat([total_df,temp_df])
    print(i)
```

```python
#define the boundary for the grid
lower_left_coordinates =[-125.000000,24.000000]
upper_right_coordinates =[-66.000000,50.000000]

filter_df = total_df[(total_df['Lontitude']>=lower_left_coordinates[0])&(total_df['Lontitude']<=upper_right_coordinates[0])&(total_df['Latitude']>=lower_left_coordinates[1])&(total_df['Latitude']<=upper_right_coordinates[1])]
# filter_df.to_csv('/home/zhoulita/scratch/covid_data_unprocessed/dataFrame/tweet_df_loc_us.csv')
filter_df
```

```python
def get_week_index(df):
    return int((df['day_index']-1)/7)
filter_df['week_index']= filter_df.apply(get_week_index,axis = 1)
filter_df
```

```python
#define the grind with the size of m_box * n_box
m_box =70
n_box =35

#use a function to get which place grid box does the tweets belong to
def get_grid_identifier(df):
    x_index = 0
    y_index = 0
    
    x_step = (upper_right_coordinates[0]-lower_left_coordinates[0])/(m_box)
    y_step = (upper_right_coordinates[1]-lower_left_coordinates[1])/(n_box)
    
    while x_index < m_box:
        if lower_left_coordinates[0]+ (x_index+1)* x_step> df['Lontitude']:
            break
        x_index = x_index+1
    
        
    while y_index < n_box:
        if lower_left_coordinates[1]+ (y_index+1)* y_step> df['Latitude']:
            break
        y_index = y_index+1
        
    return str(x_index)+"x"+str(y_index)

filter_df['identifier'] = filter_df.apply(get_grid_identifier,axis = 1)
filter_df
```

```python
week_data = filter_df[filter_df['week_index'] == 0]
grid_data_week = pd.DataFrame(week_data['identifier'].value_counts())
grid_data_week.reset_index(inplace=True)
grid_data_week.rename(columns={'identifier':'Count'},inplace=True)
grid_data_week['Count'] = grid_data_week['Count']/7
grid_data_week['Week_index'] = 0
grid_data_week
```

```python
for i in range(1,50):
    week_temp = filter_df[filter_df['week_index'] == i]
    temp = pd.DataFrame(week_temp['identifier'].value_counts())
    temp.reset_index(inplace=True)
    temp.rename(columns={'identifier':'Count'},inplace=True)
    temp['Count'] = temp['Count']/7
    temp['Week_index'] = i
    grid_data_week = pd.concat([grid_data_week,temp])
```

```python
grid_data_week.to_csv('new_processed_avg_data.csv')
```
